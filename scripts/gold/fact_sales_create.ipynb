### Initialization

```python

query = """
    SELECT 
        sd.order_number,
        pr.product_key,     -- surrogate key
        cu.customer_key,    -- surrogate key
        sd.order_date,
        sd.ship_date,
        sd.due_date,
        sd.sales_amount,
        sd.quantity,
        sd.price
    FROM catalogue_project1.silver.crm_sales_details sd
    LEFT JOIN catalogue_project1.gold.dim_products pr 
        ON sd.product_number = pr.product_number
    LEFT JOIN catalogue_project1.gold.dim_customers cu
        ON sd.customer_id = cu.customer_id
"""
```

### Check
That everything worked correctly.

```python
df = spark.sql(query)
df.limit(10).display()
```

### Write into the Gold layer delta table

```python
df.write.mode("overwrite").format("delta").saveAsTable("catalogue_project1.gold.fact_sales")
```
