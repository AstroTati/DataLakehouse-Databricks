### Initialization


```python

query = """
    SELECT 
        ROW_NUMBER() OVER (ORDER BY pn.product_start_date, pn.product_number) AS product_key, --surrogate key
        pn.product_id,
        pn.product_number,
        pn.product_name,
        pn.category_id,
        pc.category,
        pc.subcategory,
        pc.maintenance,
        pn.product_cost,
        pn.product_line,
        pn.product_start_date
    FROM catalogue_project1.silver.crm_products pn
    LEFT JOIN catalogue_project1.silver.erp_px_cat_g1v2 pc 
        ON      pn.category_id = pc.category_id
    -- WHERE pn.product_end_date IS NULL
"""
```

### Check
That everything worked correctly.

```python
df = spark.sql(query)
df.limit(10).display()
```

### Write into the Gold layer delta table

```python
df.write.mode("overwrite").format("delta").saveAsTable("catalogue_project1.gold.dim_products")
```
